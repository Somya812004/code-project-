```python

# list - mutable,ordered,allows duplicates,hetrogeneous
lst=[1,2,3,"ram"]
#print elements
print(lst[3])
print(type(lst[3]))
print(type(lst[1]))
# concatination
lst.append(5)
print(lst)
total=0
# for loop for adding two numbers
for i in lst:
    if type(i)==int:
      total = total+i
print(total)

lst.append(2)
# remove
print(lst)
print(lst.remove(5))
print(lst)
# list slicing
print(lst[0:2])
```

```python
# sets - mutable,unordered,unique elements/does not allow duplicates,hetrogenous
s = {1,2,3,'cam'}
print(s)
type(s)
#key-value pairs
s1={'name':'riku',
    'age':21}
print(s1)
print(s1['name'])

#remove function
s.remove(1)
print(s)
#add function - concat
s.add(4)
print(s)
#we cant add duplicates
s.add(4)
print(s)
```
